Import sqlalchemy and pandas.

```python
from sqlalchemy import create_engine
from sqlalchemy import inspect
import pandas as pd
import numpy as np
```

Import raw data into dataframe.

```python
engine = create_engine("sqlite:///mta_raw_2018.db")
insp = inspect(engine)
print(insp.get_table_names())
```

```python
SQLAlchemy 1: SELECT
```

```python
mta_df= pd.read_sql('SELECT CA, UNIT, SCP, STATION, DATE, TIME, DESC, ENTRIES, EXITS, DATE_TIME FROM mta_data;', engine)
mta_df
```

```python
mta_df.dtypes
```

```python
mta_df['DATE_TIME']= pd.to_datetime(mta_df['DATE']+mta_df['TIME'], format = "%m/%d/%Y%H:%M:%S")
mta_df
```

```python
mta_df.dtypes
```

Import university locations

```python
unv_loc= pd.read_sql('SELECT the_geom,LAT,LON,NAME,Technology,HOUSENUM,STREETNAME,CITY,ZIP,URL,BIN,BBL,STATION,MTA_STATION,DISTANCE,Division,Line,Station_Latitude,Station_Longitude,Route_1,Route_2,Route_3,Route_4,Route_5,Route_6,Route_7,Route_8,Route_9,Route_10,Route_11,Entrance_Type,Entry,Exit_Only,Vending,Staffing,Staff_Hours,ADA,ADA_Notes,Free_Crossover,North_South_Street,East_West_Street,Corner,Latitude,Longitude FROM university_location_mta;', engine)
unv_loc.describe()
```

```python
unv_loc.dtypes
```

Select raw data associated with universities and tag with university.

```python
unv_stn = unv_loc['MTA_STATION'].drop_duplicates().reset_index()
unv_stn
mta_df_unv = mta_df.join(unv_stn.set_index('MTA_STATION'), on='STATION')
mta_df_unv.dropna(inplace=True)
```

```python
mta_df_unv.sort_values(['CA', 'UNIT', 'SCP', 'STATION','DATE_TIME'])
```

```python
mta_df_unv['D_ENTRIES'] = mta_df_unv['ENTRIES'].diff().abs()
mta_df_unv['D_EXITS'] = mta_df_unv['EXITS'].diff().abs()
mta_df_unv
```

```python
mta_df_unv.dropna(inplace=True)
mta_df_unv
```

```python
mta_df_unv = mta_df_unv[(mta_df_unv['D_ENTRIES'] < 8000)]
mta_df_unv = mta_df_unv[(mta_df_unv['D_EXITS'] < 8000)]
mta_df_unv['TRAFFIC'] = mta_df_unv['D_ENTRIES'] + mta_df_unv['D_EXITS']
mta_df_unv.describe()
```

```python
mta_df_unv.groupby(['STATION']).TRAFFIC.sum().reset_index().sort_values(['TRAFFIC'], ascending = False).head(60)
```

```python
mta_df_unv.to_csv('mta_df_for_univ.csv', index=False)
unv_loc.to_csv('tech_univ.csv', index=False)
```
