mta_raw_data_2018 imports the mta turnstile data, does initial cleaning, and exports it to csv for use in the next analysis steps.

Import sqlalchemy and pandas.

```python
from sqlalchemy import create_engine
from sqlalchemy import inspect
import pandas as pd
```

Download from the MTA database.

```python
mta_df_210904 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_180901.txt') #get data
mta_df_210828 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_180908.txt') #get data
mta_df_210821 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_180915.txt') #get data
mta_df_210814 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_180922.txt') #get data
mta_df_210807 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_180929.txt') #get data
mta_df_210731 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_181006.txt') #get data
mta_df_210724 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_181013.txt') #get data
mta_df_210717 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_181020.txt') #get data
mta_df_210710 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_181027.txt') #get data
mta_df_210703 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_181103.txt') #get data
mta_df_210626 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_181110.txt') #get data
mta_df_210619 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_181117.txt') #get data
mta_df_210612 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_181124.txt') #get data
mta_df_210605 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_181201.txt') #get data
```

Concatenate into a single df.

```python
mta_df = pd.concat([mta_df_210904, mta_df_210828, mta_df_210821, mta_df_210814, mta_df_210807, mta_df_210731, mta_df_210724, mta_df_210717, mta_df_210710, mta_df_210703, mta_df_210626, mta_df_210619, mta_df_210612, mta_df_210605,])
mta_df.head()
```

```python
mta_df.dtypes
```

Combine Date and Time columns.

```python
mta_df['DATETIME']= pd.to_datetime(mta_df['DATE']+mta_df['TIME'], format = "%m/%d/%Y%H:%M:%S")
mta_df
```

Sort dataframe on unique turnstiles.

```python
mta_df.sort_values(['C/A', 'UNIT','SCP','STATION','DATETIME'],inplace=True)
mta_df
```

Drop duplicates.

```python
mta_df.drop_duplicates(['C/A', 'UNIT','SCP','STATION','DATETIME'])
```

Output to csv for next step in processing.

```python
mta_df.to_csv('mta_raw_2018.csv',header=False,index=False)
```
