Import sqlalchemy and pandas.

```python
from sqlalchemy import create_engine
from sqlalchemy import inspect
import math
import pandas as pd
```

Download the station dataset.

```python
stn_loc = pd.read_csv('http://web.mta.info/developers/data/nyct/subway/StationEntrances.csv') #get data
stn_loc.to_csv('stn_loc.csv')
```

Load university location data.

```python
unv_loc = pd.read_csv('COLLEGE_UNIVERSITY.csv') #get data
unv_loc.dtypes
```

```python
unv_loc
```

```python
unv_loc['STATION']='' #12
unv_loc['DISTANCE']=100 #13

for uindex, row in unv_loc.iterrows():
    
    for sindex, row in stn_loc.iterrows():
        
        x2=unv_loc.iloc[uindex,1]
        x1=stn_loc.iloc[sindex,3]
        y2=unv_loc.iloc[uindex,2]
        y1=stn_loc.iloc[sindex,4]
        
        distance = math.sqrt( (x2 - x1)**2 + (y2 - y1)**2 )
        
        if distance < unv_loc.iloc[uindex,13]:
            unv_loc.iloc[uindex,13]=distance
            unv_loc.iloc[uindex,12]=stn_loc.iloc[sindex,2]
```

```python
unv_loc
```

```python
university_location = unv_loc.join(stn_loc.set_index('Station_Name'), on='STATION')
university_location
```
